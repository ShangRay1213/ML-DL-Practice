```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Forward pass for a 6-layer net with hidden size 4096
# Weight initialization using Kaiming initialization for W1 to W6

num_layers = 6
dims = [4096] * (num_layers+1)
a = np.random.randn(16, dims[0])

outs, activations = [], []
for Din, Dout in zip(dims[:-1], dims[1:]):
  W = np.random.randn(Din, Dout) * np.sqrt(2/Din)
  out = a.dot(W)
  outs.append(out)
  a = np.maximum(out, 0)
  activations.append(a)
```

```python
def plot(lst, num_layers):
  """
  :param lst: list, storing matrices of a num_layers nueral network
  :return   : int, number of layers of a neural network
  --------------------------------------------------------
  This function extracts all the matrices in lst and plots their values on one histogram at a time.
  """
  for index, x in enumerate(lst):
    count = []
    for i in range(len(x)):
      for value in x[i]:
        count.append(value)
    plt.subplot2grid((1, num_layers), (0, index))
    plt.xlim(0, 1)
    plt.hist(count, bins=20);
```

```python
# Plot values of output of each layer
plot(outs, num_layers)
```

```python
# Plot values of activation of each layer
plot(activations, num_layers)
```
